```python
import equinox as eqx
import jax

class Model(eqx.Module):
    core: eqx.nn.Linear
    head: eqx.nn.Linear
    dropout: float

key1, key2 = jax.random.split(jax.random.key(0))
model = Model(
    core=eqx.nn.Linear(in_features=5, out_features=5, key=key1),
    head=eqx.nn.Linear(in_features=5, out_features=2, key=key2),
    dropout=0.5
)
```

```python
from refrax import focus


model = focus(model).dropout.set(0.1)
# model = focus(model).select("core", "head").bias.apply(lambda b: b + 1.0)

# We could also have used "find" for the second command
model = focus(model).find(predicate=lambda x: isinstance(x, eqx.nn.Linear)).bias.apply(lambda b: b + 1.0)

model.dropout
# 0.1

model.core.bias
# [0.646, 0.860, 0.670 , 1.277 , 0.727]

model.head.bias
# [1.634, 1.877]
```

```python
import equinox as eqx
import jax

class Model(eqx.Module):
    core: eqx.nn.Linear
    head: eqx.nn.Linear
    dropout: float

key1, key2 = jax.random.split(jax.random.key(0))
model = Model(
    core=eqx.nn.Linear(in_features=5, out_features=5, key=key1),
    head=eqx.nn.Linear(in_features=5, out_features=2, key=key2),
    dropout=0.5
)
```

```python
from refrax import focus

model = focus(model).dropout.set(0.1)
is_linear = lambda node: isinstance(node, eqx.nn.Linear)

model = focus(model).find(predicate=is_linear).bias.apply(lambda b: b + 1.0)

print(f"Dropout: {model.dropout}")
# Dropout: 0.1

print(f"Core Bias: {model.core.bias}")
# Core Bias: [0.646, 0.860, 0.670, 1.277, 0.727]

print(f"Head Bias: {model.head.bias}")
# Head Bias: [1.634, 1.877]
```
