```python
import equinox as eqx
import jax
from refrax import OpticsMixin

class Layer(eqx.Module, OpticsMixin):
    weight: jax.Array
    bias: jax.Array

class Model(eqx.Module, OpticsMixin):
    core: Layer
    head: Layer
    dropout: float

model = Model(
    core=Layer(weight=jax.numpy.ones((5, 5)), bias=jax.numpy.zeros(5)),
    head=Layer(weight=jax.numpy.ones((2, 5)), bias=jax.numpy.zeros(2)),
    dropout=0.5
)

new_model = model.at.dropout.set(0.1)
new_model = model.at.select("core", "head").bias.apply(lambda b: b + 1.0)

new_model.core.bias
# [1. 1. 1. 1. 1.]
```
